```python
from pytoda.visualizations.molecules.tmap_plotter import fareum_plot
from pytoda.visualizations.molecules.fareum_tsne_plots import *

import os
import numpy as np
```

```python
DATA_FILE = os.path.expanduser(
    '~/Box/Molecular_SysBio/data/paccmann/'
    'paccmann_affinity/all_molecules.csv'
)
ENCODE_FILE = os.path.expanduser(
    '~/Box/Molecular_SysBio/data/paccmann/'
    'paccmann_affinity/samples_latent_code.tsv'
)
```

```python

df = pd.read_csv(DATA_FILE)

continous_columns = [
    x for x in list(df.columns)
    if x not in ['Unnamed: 0', 'Unnamed: 0.1', 'source', 'SMILES', 'other']
]
```

```python
# Define visualization propertyies
def hex_to_rgb(hex_):
    hex_ = hex_.lstrip('#')
    return tuple(int(hex_[i:i+2], 16) for i in (0, 2, 4))
CMAP= {i: hex_to_rgb(x) for i,x in
       enumerate(['#AB9C55', '#404040', '#67B8F5', '#5DA87B'])}

def custom_cmap(val):
#     val = int(round(val*2))
    return CMAP[val]
import matplotlib.colors
custom_cmap = matplotlib.colors.ListedColormap(
    ['#8c8c8c', '#D74E4E', '#2BD63F', '#315936'])

SIZES_CONVERSION = {
    'unbiased': .35,
    'chembl': .25,
    'literature_candidate': .5,
    'rl_generated': .4
}
sizes = [SIZES_CONVERSION[x] for x in df.source]
```

```python
save_file_fp = 'output_tsne_fp.npy'

if os.path.exists(save_file_fp):
    output_tsne_fp = np.load(save_file_fp)
else:
    output_tsne_fp = tsne_from_fingerprints(df)
    np.save(save_file_fp, output_tsne_fp)

fareum_plot(
    df,
    output_tsne_fp,
    sizes=sizes,
    smiles_column='SMILES',
    drugs_column='other',
    categorical_columns=['source'],
    categorical_cmap=custom_cmap,
    continous_columns=continous_columns,
    background_color='#d9d9d9',
    plot_filename='tsne_fingerprints'
)

```

```python
save_file_lc = 'output_tsne_lc.p30.npy'

if os.path.exists(save_file_lc):
    output_tsne_lc = np.load(save_file_lc)
else:
    output_tsne_lc = tsne_from_properties_file(
        df, ENCODE_FILE,
        tsne_params = {'number_components': 3,
                       'perplexity': 70})
    np.save(save_file_lc, output_tsne_lc)

fareum_plot(
    df,
    output_tsne_lc,
    sizes=sizes,
    smiles_column='SMILES',
    drugs_column='other',
    categorical_columns=['source'],
    categorical_cmap=custom_cmap,
    continous_columns=continous_columns,
    background_color='#d9d9d9',
    plot_filename='tsne_latentcode'
)
```

```python
save_file_lc = 'output_umap_lc.n80.npy'

if os.path.exists(save_file_lc) and False:
    output_tsne_lc = np.load(save_file_lc)
else:
    output_tsne_lc = umap_from_properties_file(
        df,
        ENCODE_FILE,
        umap_params = {'number_components': 3,
                       'n_neighbors': 80,
                        'min_dist': 5}      
    )
    np.save(save_file_lc, output_tsne_lc)

fareum_plot(
    df,
    output_tsne_lc,
    sizes=sizes,
    smiles_column='SMILES',
    drugs_column='other',
    categorical_columns=['source'],
    categorical_cmap=custom_cmap,
    continous_columns=continous_columns,
    background_color='#d9d9d9',
    plot_filename='umap_latentcode'
)
```

```python
save_file_lc = 'output_umap_fp.n100.npy'

if os.path.exists(save_file_lc) and False:
    output_tsne_lc = np.load(save_file_lc)
else:
    output_tsne_lc = umap_from_fingerprints(
        df,
        umap_params = {'number_components': 3,
                       'n_neighbors': 100,
                        'min_dist': 5}        
    )
    np.save(save_file_lc, output_tsne_lc)

fareum_plot(
    df,
    output_tsne_lc,
    sizes=sizes,
    smiles_column='SMILES',
    drugs_column='other',
    categorical_columns=['source'],
    categorical_cmap=custom_cmap,
    continous_columns=continous_columns,
    background_color='#d9d9d9',
    plot_filename='umap_fingerprints'
)
```
